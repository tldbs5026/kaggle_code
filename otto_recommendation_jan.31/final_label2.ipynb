```python
import pandas as pd
```

```python
fm_submission = pd.read_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/fm_submission1.csv')
w2v_submission = pd.read_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/w2v1_sub.csv')
```

```python
sub = pd.merge(fm_submission, w2v_submission, on=['session_type', 'labels'], how='left')
```

```python
sub = sub[2_000_000:4_000_000]
```

```python
w_labels = w2v_submission['labels']
w_labels
```

```python
w_labels2 = w_labels[2000000:4000000]
```

```python
w_labels2 = w_labels2.str.split(' ')
```

```python
w_labels2
```

```python
f_labels= fm_submission['labels']
f_labels
```

```python
f_labels2 =  f_labels[2000000:4000000]
```

```python
f_labels2 = f_labels2.str.split(' ')
```

```python
f_labels2
```

```python
sub_lists = []
for i in range(len(f_labels2)) :
  sub_list = [item for sublist1 in zip(w_labels2[2000000+i], f_labels2[2000000+i]) for item in sublist1]
  sub_list = list(dict.fromkeys(sub_list))
  sub_lists.append(sub_list[:20])
sub_lists[:5]
```

```python
labels_as_strings = [' '.join([str(l) for l in lls]) for lls in sub_lists]
submission = pd.DataFrame(data={'session_type': sub['session_type'], 'labels': labels_as_strings})

submission = submission.sort_values('session_type', axis=0, ascending=True)
```

```python
submission.to_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/submission2.csv', index=False)
```
