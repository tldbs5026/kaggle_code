```python
import pandas as pd
```

```python
fm_submission = pd.read_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/fm_submission1.csv')
w2v_submission = pd.read_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/w2v1_sub.csv')
```

```python
sub = pd.merge(fm_submission, w2v_submission, on=['session_type', 'labels'], how='left')
```

```python
sub = sub[4_000_000:]
```

```python
w_labels = w2v_submission['labels']
w_labels
```

```python
w_labels3 = w_labels[4000000:]
```

```python
w_labels3 = w_labels3.str.split(' ')
```

```python
f_labels= fm_submission['labels']
f_labels
```

```python
f_labels3 = f_labels[4000000:]
```

```python
f_labels3 = f_labels3.str.split(' ')
```

```python
sub_lists3 = []
for i in range(len(sub)) :
    sub_list = [item for sublist3 in zip(w_labels3[4000000+i], f_labels3[4000000+i]) for item in sublist3]
    sub_list = list(dict.fromkeys(sub_list))
    sub_lists3.append(sub_list[:20])
sub_lists3[:5]
```

```python
labels_as_strings = [' '.join([str(l) for l in lls]) for lls in sub_lists3]
submission = pd.DataFrame(data={'session_type': sub['session_type'], 'labels': labels_as_strings})

submission = submission.sort_values('session_type', axis=0, ascending=True)
```

```python
submission.to_csv('/content/drive/MyDrive/주피터_대피소/kaggle&github/Kaggle/참가중대회/otto~jan31,2023/submission3.csv', index=False)
```
