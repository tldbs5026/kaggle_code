```python
data=pd.concat([train,test],join = 'outer', axis=0)
```

```python
data.head()
```

```python
y_data=train.pop('Survived')

x_data=train['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
# x_data=dataset
# y_data=dataset.pop('Survived')


print(x_data.shape)
print(y_data.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_data,y_data, test_size=0.2, random_state=42, stratify=y_data)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

#model evaluation comparison

```python
from lazypredict.Supervised import LazyClassifier
#지도학습중 LazyClassifier를 사용


clf=LazyClassifier(verbose=0, predictions=True)

models, predictions = clf.fit(x_train, x_test, y_train, y_test)

models
```

```python
predictions
```

classification report by models

```python
from sklearn.metrics import classification_report

for model_name in predictions.columns.tolist() :
  print(f'{model_name}')
  print(classification_report(y_test, predictions[model_name]))



```

```python
import numpy as np

from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipe=Pipeline(steps=[
    ('preprocessor', StandardScaler()),
    ('classifier', SVC())
])

pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)

y_acc=y_pred.best_estimators_
y_pred.best_score
```

checking hyperparameters of model

```python
pipe['classifier'].get_params()
```

#Scikit-Optimize

scikit-learn의 머신러닝 모델들의 파라미터 자동 튜닝

```python
from skopt import BayesSearchCV


pipe=Pipeline(steps = [
    ('preprocessor', StandardScaler()),
    ('classifier', SVC())
])

opt=BayesSearchCV(pipe,
                  {
                      'classifier__kernel' : ['linear','rbf','poly'],
                   'classiifer__C' : (1e-1,1e+1, 'log-uniform'),
                   'classifier__gamma' : (1e-6,1e+1, 'log-uniform'),
                   'classifier__degree' : (1,8)
                  },
                  n_iter=8,
                  cv=3)


opt.fit(x_train,y_train)
y_pred=opt.predict(x_test)

print('val. score : {}'.foramt(opt.best_score_))
print('test. score : {}'.foramt(opt.score(x_test,y_test)))

```

checking which paramter is best

```python
opt.best_params_
```
